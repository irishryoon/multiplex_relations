```julia
using Distances
using Plots
using Ripserer
using Eirene
using PersistenceDiagrams
using Images
using LinearAlgebra
using JLD
using Random
using JLD2
include("compute_ph2.jl")
workdir = pwd()
```

```julia
# VR H0 Tumour persistence Images
location_tumour_h0 = "\\persistence_diagrams\\vr\\tumour_h0\\"
all_diags=[]
files_tumour_h0 = readdir(workdir*location_tumour_h0)
for file in files_tumour_h0
    diagram = load_diagram(workdir*location_tumour_h0*file,0)
    append!(all_diags,[diagram])
end
pi_location_tumour_h0 = "persistence_images\\vr\\tumour_h0\\"
persistence_image = PersistenceImage(all_diags,size=20,sigma=1)
for file in files_tumour_h0
    id,time=parse_file(file)
    A = id*"_"*time
    name = A*"_tumour_h0_img.jld2"
    path = pi_location_tumour_h0*name
    diagram = load_diagram(workdir*location_tumour_h0*file,0)
    img = persistence_image(diagram)
    save_object(path,img)
end
```
